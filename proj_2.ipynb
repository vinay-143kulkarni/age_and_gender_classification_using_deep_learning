```python
import tensorflow as tf
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
## Convolution(256 feature detector of dimension 3 by 3), input shape 3 layer for color image)
classifier.add(Conv2D(256,(7,7),input_shape = (256,256,3), activation = 'relu'))
## MaxPooling
classifier.add(MaxPooling2D(pool_size = (2,2)))

## Add another layer
classifier.add(Conv2D(256,(5,5), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))

## Add another layer
classifier.add(Conv2D(256,(3,3), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))
```

```python
classifier.add(Flatten())
```

```python
## Fully connected ANN, Hidden ANN and output layer
classifier.add(Dense(units = 500, activation = 'relu'))
#classifier.add(Dense(units = 200, activation = 'relu'))
#classifier.add(Dense(units = 100, activation = 'relu'))
classifier.add(Dropout(0.2))
```

```python
#output layer
classifier.add(Dense(units = 14, activation = 'softmax'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
classifier.load_weights('final_weights_v.h5')
```

```python
from keras.preprocessing import image
test_image= image.load_img('coarse_tilt_aligned_face.723.12059004943_860ba14a52_o.jpg'
                           ,target_size =(256,256))
```
